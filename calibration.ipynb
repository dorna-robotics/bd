# BD Calibration

Probe calibration for the **bd** falcon bench, on the calibration scene
(`scene/layout_calibration.j2`): the two tool racks hold `probe_rod`
(144 mm, simple mode) and `probe_vertical` (192 mm, precise mode) instead
of the production gripper / pipettor.

**Recipes are the production `recipes.j2`** — calibration must use the
exact recipe params production uses, because the calibration point is
keyed by `calibration_name` (component + left_approach + base_distance +
rail_step + rail_span). Using the production handles guarantees production
`interpolate()` finds what we calibrate here. The handles are named by
their production role, so the calibration-target mapping is:

| handle      | component            | calibrate            |
|-------------|----------------------|----------------------|
| `tip_rack`  | `adapter_sbs_0`      | sbs 0 (simple)       |
| `waste_bin` | `adapter_sbs_1`      | sbs 1 (simple)       |
| `falcon_rack`   | `adapter_plate_falcon_15ml_1` | falcon adapter (simple) |
| `falcon_pipette`| `adapter_plate_falcon_15ml_1` | falcon adapter (simple) |
| `decapper`      | `decapper_1`         | decapper (simple)       |
| `capholder`     | `adapter_capholder_falcon_15ml_1` | cap holder (simple)     |
| `gripper`   | `tool_rack_144mm_1`  | tool changer (precise, probe_rod's rack)      |
| `pipettor`  | `tool_rack_192mm_1`  | tool changer (precise, probe_vertical's rack) |

Flow:
1. **Scene initialization** — build the workspace, turn sim on, mount a probe.
2. **Calibration 1** — the two sbs adapters, one cell per clb anchor,
   `mode="simple"`, `tool_anchor="tcp_0"` (uses `probe_rod`).
3. **Calibration 2** — the falcon adapter, both recipes (`falcon_rack`,
   `falcon_pipette`), one cell per clb anchor, `mode="simple"` (uses
   `probe_rod`).
4. **Calibration 3** — the decapper, one cell per clb anchor,
   `mode="simple"`, `tool_anchor="tcp_0"` (uses `probe_rod`).
5. **Calibration 4** — the cap holder, one cell per clb anchor,
   `mode="simple"`, `tool_anchor="tcp_0"` (uses `probe_rod`).
6. **Tool-changer calibration** — the two tool racks, `mode="precise"`
   (uses `probe_vertical`).


## Scene initialization

```python
from pathlib import Path

from workspace.workspace import Workspace
from workspace.bt.launcher import load_recipes

# Run this notebook from the bd project directory.
BASE = Path.cwd()

scene = [
    str(BASE / "scene" / "core_500.j2"),
    str(BASE / "scene" / "layout_calibration.j2"),
]

workspace = Workspace(config_path=scene, port=5000)
core = workspace.components["core"]
# Production recipes — same handles/params production uses, so calibration
# is keyed to exactly what interpolate() looks up at run time.
rcp  = load_recipes(workspace, core, BASE / "recipes.j2")
```

### Turn simulation on

```python
# Route every recipe / motion call through the SimulationAPI (no hardware,
# no bus claim). Flip to False for a real calibration run.
core.simulation(False)
```

### Tool change

`calibrate()` needs a probe mounted on the robot (`core.current_tool()`
must be set). Mount / park probes here, or attach by hand in the viewer.

```python
# probe_rod      -> 144 mm rack -> handle "gripper"  -> SIMPLE mode (adapter_sbs_*)
# probe_vertical -> 192 mm rack -> handle "pipettor" -> PRECISE mode (tool changers)

# mount probe_rod for the adapter_sbs simple calibrations:
rcp["gripper"].pick()

# ...when done, park it and mount probe_vertical for the tool changers:
# rcp["gripper"].place()
# rcp["pipettor"].pick()

tool = core.current_tool()
print("current tool:", tool.name if tool else None)
```
